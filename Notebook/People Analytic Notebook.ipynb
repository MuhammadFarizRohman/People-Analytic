```python
#Visualisasi Faktor mempengaruhi kepuasan/job_satisfaction
numerical_df = df.select_dtypes(include=['int64', 'float64'])
corr = numerical_df.corr()
plt.figure(figsize=(14, 10))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Korelasi Heatmap job_santisfaction terhadap semua variabel", fontsize=14)
plt.show()
```

```python
#EDA Visualisasi Barplot rata rata kepuasan Pekerja berdasarkan departement

ax = df.groupby("dept")["job_satisfaction"].mean().plot(kind='bar', figsize=(12,6))

for p in ax.patches:
    ax.annotate(
        format(p.get_height(), ".2f"),
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center', va='bottom', fontsize=10
    )

plt.title("Rata rata kepuasan pekerja")
plt.ylabel("Job Satisfaction")
plt.xlabel("Department")
plt.show()
```

Yang Saya Tangkap ialah memiliki kedataran pada tiap departement yang artinya tidak ada gap sama sekali antar departement , distribusi ini disebut uniform

```python
#EDA Visualisasi Barplot rata rata kepuasan Pekerja berdasarkan Lvl Education
ax = df.groupby("edu_level")["job_satisfaction"].mean().plot(kind='bar', figsize=(12,6))

for p in ax.patches:
    ax.annotate(
        format(p.get_height(), ".2f"),
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center', va='bottom', fontsize=10
    )

plt.title("Rata rata kepuasan pekerja berdasarkan lvl edukasi")
plt.ylabel("Job Satisfaction")
plt.xlabel("Level Education")
plt.show()
```

```python
#Membuat Segmentasi berdasarkan kepuasan bekerja
df['satisfaction_segment'] = pd.qcut(df['job_satisfaction'], q=3, labels=['Kecil', 'Sedang', 'Tinggi'])
#menampilkan data kepuasan per departement
df.groupby(['dept', 'satisfaction_segment']).size().groupby(level=0).apply(lambda x: (x / x.sum()) * 100)
```
